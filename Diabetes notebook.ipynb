```python
import pandas as pd
import numpy as np
```

# Reading training data

```python
df = pd.read_csv(r"C:\Users\Saideep\Downloads\0000000000002329_training_diabetes_x_y_train.csv");
df.shape
```

```python
df.describe()
```

```python
X_train = np.array(df.iloc[:,0:10])
Y_train = np.array(df.iloc[:,10])
X_train,Y_train
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
algo = LinearRegression()
```

```python
algo.fit(X_train,Y_train)
algo.coef_,algo.intercept_
```

# Reading testing data

```python
df = pd.read_csv(r"C:\Users\Saideep\Downloads\0000000000002329_test_diabetes_x_test.csv")
X_test = np.array(df.iloc[:,0:10])
X_test.shape
```

```python
Y_predict = algo.predict(X_test)
Y_predict_rounded = np.around(Y_predict, decimals=5)
Y_predict_rounded
```
